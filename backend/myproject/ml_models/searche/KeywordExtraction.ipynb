```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
```

```python
with open('test1.txt', 'r', encoding="utf8") as file:
    text2 = file.read().replace('\n', '')
```

```python
print(summarize(text2, ratio=0.1))
```

```python
print(summarize(text2, word_count=100))
```

```python
print(keywords(text2, words=20, lemmatize=True))
```
